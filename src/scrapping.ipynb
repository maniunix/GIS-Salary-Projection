```python
# !pip install selenium
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
from selenium import webdriver
import time
import pandas as pd
from selenium.webdriver.common.by import By
```

__Defined Class and Methods__ 

```python
class scrapper:
##### Scrapper Accepts the #####
##### ** Keyword ** which will be JOB ROLE #####
##### ** location ** which will be location of interest i.e. India #####
##### ** Page ** Total number of Pages you would like to scrap #####

    def __init__(self, keyword, location, page):
        self.keyword = keyword
        self.location = location
        self.page = page

    def getJobDesc(self):
        path = "E:/Workspace/gis_salary_proj/chromedriver"
        options = webdriver.ChromeOptions()
        driver = webdriver.Chrome(executable_path=path, options=options)
        driver.set_window_size(1120, 1000)
        job_desc = []
        url = f"https://in.indeed.com/jobs?q={self.keyword}&l={self.location}&start={self.page}0&pp=gQAPAAABhjpfH0EAAAAB-HMzigAdAQA9j26Kn2PmZeS6I04EDTkITNq89HjrpZMf1aYAAA&vjk=f71c00e996a2405e"
        driver.get(url)
        driver.implicitly_wait(20)
        try:
            job_description = driver.find_element(By.ID,"mosaic-provider-jobcards").text
            job_desc.append(job_description)
        except ElementClickInterceptedException as ee:
            print(ee)
        return job_description
    
    def cleanData(self):
        job_description = [i.split("\n") for i in self.getJobDesc().split("days ago")]
        job_cards = [i[1:-2] for i in job_description[1:]]
        job_cards.append(job_description[0])
        return job_cards
```

```python
jobs = []
for i in range(1,9):
    foo = scrapper(location="India",keyword="GIS Developer", page=i)
    jobs.append(foo.cleanData())
```
